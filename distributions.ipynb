```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
path_to_json = "IUM22Z_Zad_02_01_v2/"
json_files = []
for j in os.listdir(path_to_json):
    json_files.append(path_to_json + j)
json_files
```

```python
data = []
for j in json_files:
    df = pd.read_json(j, lines=True)
    data.append(df)
```

```python
data[4]
```

```python
data[4].hist(column= ["popularity"], figsize=(16, 9))
plt.show()
data[4]["popularity"].describe()
```

```python
data[4].hist(column= ["duration_ms"], figsize=(16, 9), range=[0, 500000])
plt.show()
data[4]["duration_ms"].describe()
```
